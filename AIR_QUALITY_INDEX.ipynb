Finding Null Values

```python
data.isnull().sum()
```

Filling null values

```python

columns_to_fill = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene', 'AQI']
for column in columns_to_fill:
    data[column] = data[column].fillna(data[column].mean())
```

Check if there any null value after filling nan

```python
data.isnull().sum()
```

Ensure 'date' is a datetime type

```python
data['Date'] = pd.to_datetime(data['Date'])
```

add row for year and month

```python
data['Month'] = data['Date'].dt.month
data['Year'] = data['Date'].dt.year
```

1.Analysis of AQI Bucket

Counting AQI Bucket

```python
data['AQI_Bucket'].value_counts()
```

Plotting Graph for AQI Bucket

```python
aqi_buckets_count = data['AQI_Bucket'].value_counts()
plt.figure(figsize=(10, 6))
aqi_buckets_count.plot(kind='bar', color='skyblue')
plt.title('Count of AQI Buckets')
plt.xlabel('AQI Bucket')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

2.Analysis of AQI

Plotting graph for AQI

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['AQI'], bins=30, kde=True, color='blue')
plt.title('Distribution of Air Quality Index (AQI)')
plt.xlabel('AQI')
plt.ylabel('Frequency')
plt.show()
```

Group data by month and year for AQI

```python
AQI_trends = data.groupby('Month'and'Year')['AQI'].mean().reset_index()
AQI_trends = pd.DataFrame(data)
```

Plotting Graph for Seasonal Trends in AQI Over the Years

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Month', y='AQI', hue='Year', data=AQI_trends, palette='Set1')
plt.title('Seasonal Trends in AQI Over the Years')
plt.xlabel('Month')
plt.ylabel('AQI')
plt.legend(title='Year', loc='lower right')
plt.show()
```

Plot seasonal trends in AQI over the months

```python
seasonal_trends = data.groupby('Month' and 'Year')['AQI'].mean()


plt.figure(figsize=(14, 6))
plt.plot(seasonal_trends.index, seasonal_trends.values, marker='o', linestyle='-', color='green')
plt.title('seasonal trends in AQI over the months')
plt.xlabel('Month')
plt.ylabel('AQI')
plt.grid(True)
plt.show()
```

3.Pollution spikes

Group data by City for AQI

```python
city_trends = data.groupby('City')['AQI'].mean()
city_trends = pd.DataFrame(data)
```

Plot AQI over time to identify spikes

```python
### Identify Pollution Spikes and Potential Causes ###
plt.figure(figsize=(14, 7))
sns.lineplot(x='Date', y='AQI', data=city_trends, hue='City')
plt.title('Daily AQI Over Time Across Cities')
plt.xlabel('Date')
plt.ylabel('AQI')
plt.legend(title='City', loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

4.AQI by Cities and their Comparison

For specific city analysis

```python
city_name = 'Delhi'  # Change this to any city in your dataset
city_data = data[data['City'] == city_name]
```

Plot pollution levels for Delhi

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='AQI', data=city_data)
plt.title('AQI in', city_name, 'Over Time')
plt.xlabel('Date')
plt.ylabel('AQI')
plt.show()
```

Compare AQI Between Cities

Calculate the average AQI for each city

```python
city_comparison = data.groupby('City')['AQI'].mean().reset_index()
# Sort by AQI values
city_comparison = city_comparison.sort_values(by='AQI', ascending=False)
```

Plot Graph for Comparison b/w  AQI of Cities

```python
plt.figure(figsize=(14, 7))
sns.barplot(x='AQI', y='City', data=city_comparison, palette='coolwarm', hue='AQI')
plt.title('Comparison b/w  AQI of Cities')
plt.xlabel('AQI')
plt.ylabel('City')
plt.legend(title='AQI')  # Optional: Add a legend title
plt.show()
```

6.Correlation between Pollutants

Compute the correlation matrix for pollutants

```python
pollutant_columns = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene', 'AQI']
correlation_matrix = data[pollutant_columns].corr()
```

Plot heatmap for the correlation between pollutants

```python
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Between Pollutants')
plt.show()
```
